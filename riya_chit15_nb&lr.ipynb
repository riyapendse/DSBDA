```python
def outliers(df_out, drop=False):
    for feature in df_out.columns:
        feature_data = df_out[feature]
        q1 = np.percentile(feature_data,25.)
        q3 = np.percentile(feature_data,75.)
        iqr = q3-1
        step = iqr*1.5
        outliers = feature_data[~((feature_data>=q1-step)&(feature_data<=q3+step))].index.tolist()
        if not drop:
            print(f"feature: {feature} , No. of outliers: {len(outliers)}")
        if drop:
            df.drop(outliers, inplace=True,errors='ignore')
            print(f"outliers from {feature} are removed.")
```

### Naive Bayes

```python
from sklearn.preprocessing import LabelEncoder
labelenc = LabelEncoder()
for feature in df.columns:
    df[feature] = labelenc.fit_transform(df[feature])
```

```python
X = df.drop(['class'],axis=1).values
y = df['class']
```

```python
from sklearn.naive_bayes import MultinomialNB
multinom = MultinomialNB()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=41)
```

```python
multinom.fit(X_train,y_train)
```

```python
y_pred = multinom.predict(X_test)
```

```python
from sklearn import metrics
print("1. Accuracy Score", metrics.accuracy_score(y_test,y_pred)*100)
print("2. Precision Score:",metrics.precision_score(y_test, y_pred))
print("3. Recall Score:", metrics.recall_score(y_test, y_pred))
print("4. f1 Score:", metrics.f1_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred1=logistic.predict(X_test)
```

```python
from sklearn import metrics
print("1. Accuracy Score", metrics.accuracy_score(y_test,y_pred1)*100)
print("2. Precision Score:",metrics.precision_score(y_test, y_pred1))
print("3. Recall Score:", metrics.recall_score(y_test, y_pred1))
print("4. f1 Score:", metrics.f1_score(y_test, y_pred1))
```
