```python
import matplotlib.pyplot as plt
import seaborn as sns
variables = ['age','sex']
for var in variables:
    df[var].hist(bins=10, figsize=(10,6))
    print(f"for {var}")
    plt.tight_layout()
    plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(10,6))
    df[var].value_counts().plot(kind='bar')
    plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(10,6))
    plt.plot(df[var])
    plt.show()
```

```python
sns.scatterplot(x="age",y="sex",data=df)
plt.show()
```
