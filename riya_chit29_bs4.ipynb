```python
import bs4
import requests
```

```python
myreq = requests.get('https://www.flipkart.com/oneplus-nord-2t-5g-jade-fog-128-gb/p/itm31ac32e308282?pid=MOBGHBZHFAUYKH3U&lid=LSTMOBGHBZHFAUYKH3UHZ1SIE&marketplace=FLIPKART&q=one+plus+mobile&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&fm=Search&iid=8343cf38-e14a-4f85-9d5d-bd9c13d9c009.MOBGHBZHFAUYKH3U.SEARCH&ppt=sp&ppn=sp&ssid=s8zpw4rcls0000001685298692355&qH=8de9fb854f805dea')
myreq
```

```python
myreq.content
```

```python
soup = bs4.BeautifulSoup(myreq.text)
soup
```

```python
reviews = soup.findAll('div',{'class':'t-ZTKy'})
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
avg_rating = soup.find('div',{'class':'_2d4LTz'}).get_text()
avg_rating
```

```python
indi_ratings = soup.findAll('div',{'class':'_3LWZlK _1BLPMq'})
for indi in indi_ratings:
    print(indi.get_text()+"\n")
```

```python
customers = soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
for cust in customers:
    print(cust.get_text()+"\n")
```

```python
likes = soup.findAll('span',{'class':'_3c3Px5'})
for like in likes:
    print(like.get_text()+"\n")
```

```python
questions = soup.findAll('div',{'class':'_1xR0kG _3cziW5'})
for qs in questions:
    print(qs.get_text()+"\n")
```

```python
price = soup.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text()
price
```
